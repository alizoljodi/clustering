```python

import numpy as np
import torch
import torch.nn as nn
import argparse
import os
import random
import time
import hubconf  # noqa: F401
import copy
from quant import (
    block_reconstruction,
    layer_reconstruction,
    BaseQuantBlock,
    QuantModule,
    QuantModel,
    set_weight_quantize_params,
)
from data.imagenet import build_imagenet_data


def seed_all(seed=1029):
    random.seed(seed)
    os.environ['PYTHONHASHSEED'] = str(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    torch.cuda.manual_seed(seed)
    torch.cuda.manual_seed_all(seed)  # if you are using multi-GPU.
    torch.backends.cudnn.benchmark = False
    torch.backends.cudnn.deterministic = True


class AverageMeter(object):
    """Computes and stores the average and current value"""
    def __init__(self, name, fmt=':f'):
        self.name = name
        self.fmt = fmt
        self.reset()

    def reset(self):
        self.val = 0
        self.avg = 0
        self.sum = 0
        self.count = 0

    def update(self, val, n=1):
        self.val = val
        self.sum += val * n
        self.count += n
        self.avg = self.sum / self.count

    def __str__(self):
        fmtstr = '{name} {val' + self.fmt + '} ({avg' + self.fmt + '})'
        return fmtstr.format(**self.__dict__)


class ProgressMeter(object):
    def __init__(self, num_batches, meters, prefix=""):
        self.batch_fmtstr = self._get_batch_fmtstr(num_batches)
        self.meters = meters
        self.prefix = prefix

    def display(self, batch):
        entries = [self.prefix + self.batch_fmtstr.format(batch)]
        entries += [str(meter) for meter in self.meters]
        print('\t'.join(entries))

    def _get_batch_fmtstr(self, num_batches):
        num_digits = len(str(num_batches // 1))
        fmt = '{:' + str(num_digits) + 'd}'
        return '[' + fmt + '/' + fmt.format(num_batches) + ']'


def accuracy(output, target, topk=(1,)):
    """Computes the accuracy over the k top predictions for the specified values of k"""
    with torch.no_grad():
        maxk = max(topk)
        batch_size = target.size(0)

        _, pred = output.topk(maxk, 1, True, True)
        pred = pred.t()
        correct = pred.eq(target.view(1, -1).expand_as(pred))

        res = []
        for k in topk:
            correct_k = correct[:k].reshape(-1).float().sum(0, keepdim=True)
            res.append(correct_k.mul_(100.0 / batch_size))
        return res

@torch.no_grad()
def validate_model(val_loader, model, device=None, print_freq=100):
    if device is None:
        device = next(model.parameters()).device
    else:
        model.to(device)
    batch_time = AverageMeter('Time', ':6.3f')
    top1 = AverageMeter('Acc@1', ':6.2f')
    top5 = AverageMeter('Acc@5', ':6.2f')
    progress = ProgressMeter(
        len(val_loader),
        [batch_time, top1, top5],
        prefix='Test: ')

    # switch to evaluate mode
    model.eval()

    end = time.time()
    for i, (images, target) in enumerate(val_loader):
        images = images.to(device)
        target = target.to(device)

        # compute output
        output = model(images)

        # measure accuracy and record loss
        acc1, acc5 = accuracy(output, target, topk=(1, 5))
        top1.update(acc1[0], images.size(0))
        top5.update(acc5[0], images.size(0))

        # measure elapsed time
        batch_time.update(time.time() - end)
        end = time.time()

        if i % print_freq == 0:
            progress.display(i)

    print(' * Acc@1 {top1.avg:.3f} Acc@5 {top5.avg:.3f}'.format(top1=top1, top5=top5))

    return top1.avg

def get_train_samples(train_loader, num_samples):
    train_data, target = [], []
    for batch in train_loader:
        train_data.append(batch[0])
        target.append(batch[1])
        if len(train_data) * batch[0].size(0) >= num_samples:
            break
    return torch.cat(train_data, dim=0)[:num_samples], torch.cat(target, dim=0)[:num_samples]


if __name__ == '__main__':

    parser = argparse.ArgumentParser(description='running parameters',
                                     formatter_class=argparse.ArgumentDefaultsHelpFormatter)
    # general parameters for data and model
    parser.add_argument('--seed', default=1005, type=int, help='random seed for results reproduction')
    parser.add_argument('--arch', default='resnet18', type=str, help='model name',
                        choices=['resnet18', 'resnet50', 'mobilenetv2', 'regnetx_600m', 'regnetx_3200m', 'mnasnet'])
    parser.add_argument('--batch_size', default=64, type=int, help='mini-batch size for data loader')
    parser.add_argument('--workers', default=4, type=int, help='number of workers for data loader')
    parser.add_argument('--data_path', default='/mimer/NOBACKUP/groups/naiss2025-22-91/imagenet', type=str, help='path to ImageNet data')

    # quantization parameters
    parser.add_argument('--n_bits_w', default=2, type=int, help='bitwidth for weight quantization')
    parser.add_argument('--channel_wise', default=True, help='apply channel_wise quantization for weights')
    parser.add_argument('--n_bits_a', default=2, type=int, help='bitwidth for activation quantization')
    parser.add_argument('--disable_8bit_head_stem', action='store_true')

    # weight calibration parameters
    parser.add_argument('--num_samples', default=1024, type=int, help='size of the calibration dataset')
    parser.add_argument('--iters_w', default=20000, type=int, help='number of iteration for adaround')
    parser.add_argument('--weight', default=0.01, type=float, help='weight of rounding cost vs the reconstruction loss.')
    parser.add_argument('--keep_cpu', action='store_true', help='keep the calibration data on cpu')

    parser.add_argument('--b_start', default=20, type=int, help='temperature at the beginning of calibration')
    parser.add_argument('--b_end', default=2, type=int, help='temperature at the end of calibration')
    parser.add_argument('--warmup', default=0.2, type=float, help='in the warmup period no regularization is applied')

    # activation calibration parameters
    parser.add_argument('--lr', default=4e-5, type=float, help='learning rate for LSQ')

    parser.add_argument('--init_wmode', default='mse', type=str, choices=['minmax', 'mse', 'minmax_scale'],
                        help='init opt mode for weight')
    parser.add_argument('--init_amode', default='mse', type=str, choices=['minmax', 'mse', 'minmax_scale'],
                        help='init opt mode for activation')

    parser.add_argument('--prob', default=0.5, type=float)
    parser.add_argument('--input_prob', default=0.5, type=float)
    parser.add_argument('--lamb_r', default=0.1, type=float, help='hyper-parameter for regularization')
    parser.add_argument('--T', default=4.0, type=float, help='temperature coefficient for KL divergence')
    parser.add_argument('--bn_lr', default=1e-3, type=float, help='learning rate for DC')
    parser.add_argument('--lamb_c', default=0.02, type=float, help='hyper-parameter for DC')
    args, unknown = parser.parse_known_args()


    seed_all(args.seed)
    # build imagenet data loader
    train_loader, test_loader = build_imagenet_data(batch_size=args.batch_size, workers=args.workers,
                                                    data_path=args.data_path)
    # load model
    cnn = eval('hubconf.{}(pretrained=True)'.format(args.arch))
    cnn.cuda()
    cnn.eval()
    fp_model = copy.deepcopy(cnn)
    fp_model.cuda()
    fp_model.eval()

    # build quantization parameters
    wq_params = {'n_bits': args.n_bits_w, 'channel_wise': args.channel_wise, 'scale_method': args.init_wmode}
    aq_params = {'n_bits': args.n_bits_a, 'channel_wise': False, 'scale_method': args.init_amode,
                 'leaf_param': True, 'prob': args.prob}

    fp_model = QuantModel(model=fp_model, weight_quant_params=wq_params, act_quant_params=aq_params, is_fusing=False)
    fp_model.cuda()
    fp_model.eval()
    fp_model.set_quant_state(False, False)
    qnn = QuantModel(model=cnn, weight_quant_params=wq_params, act_quant_params=aq_params)
    qnn.cuda()
    qnn.eval()
    if not args.disable_8bit_head_stem:
        print('Setting the first and the last layer to 8-bit')
        qnn.set_first_last_layer_to_8bit()

    qnn.disable_network_output_quantization()
    print('the quantized model is below!')
    print(qnn)
    cali_data, cali_target = get_train_samples(train_loader, num_samples=args.num_samples)
    device = next(qnn.parameters()).device

    # Kwargs for weight rounding calibration
    kwargs = dict(cali_data=cali_data, iters=args.iters_w, weight=args.weight,
                b_range=(args.b_start, args.b_end), warmup=args.warmup, opt_mode='mse',
                lr=args.lr, input_prob=args.input_prob, keep_gpu=not args.keep_cpu, 
                lamb_r=args.lamb_r, T=args.T, bn_lr=args.bn_lr, lamb_c=args.lamb_c)


    '''init weight quantizer'''
    set_weight_quantize_params(qnn)

    def set_weight_act_quantize_params(module, fp_module):
        if isinstance(module, QuantModule):
            layer_reconstruction(qnn, fp_model, module, fp_module, **kwargs)
        elif isinstance(module, BaseQuantBlock):
            block_reconstruction(qnn, fp_model, module, fp_module, **kwargs)
        else:
            raise NotImplementedError
    def recon_model(model: nn.Module, fp_model: nn.Module):
        """
        Block reconstruction. For the first and last layers, we can only apply layer reconstruction.
        """
        for (name, module), (_, fp_module) in zip(model.named_children(), fp_model.named_children()):
            if isinstance(module, QuantModule):
                print('Reconstruction for layer {}'.format(name))
                set_weight_act_quantize_params(module, fp_module)
            elif isinstance(module, BaseQuantBlock):
                print('Reconstruction for block {}'.format(name))
                set_weight_act_quantize_params(module, fp_module)
            else:
                recon_model(module, fp_module)
    # Start calibration
    recon_model(qnn, fp_model)

    qnn.set_quant_state(weight_quant=True, act_quant=True)
    print('Full quantization (W{}A{}) accuracy: {}'.format(args.n_bits_w, args.n_bits_a,
                                                           validate_model(test_loader, qnn)))
```

```python
def hash_key(q_logits, topk=5, num_bins=64):
    """
    Creates a discretized key using the top-k logits.
    """
    top_vals, _ = q_logits.topk(topk, dim=-1)  # shape: [B, topk]
    min_val = top_vals.min()
    max_val = top_vals.max()
    step = (max_val - min_val + 1e-6) / num_bins

    bins = ((top_vals - min_val) / step).long().clamp(0, num_bins - 1)
    return [tuple(b.cpu().tolist()) for b in bins]
```

```python
from collections import defaultdict

def build_lut(q_model, fp_model, dataloader, device, topk=5, num_bins=64):
    lut = defaultdict(list)

    q_model.eval()
    fp_model.eval()

    with torch.no_grad():
        for images, _ in dataloader:
            images = images.to(device)
            q_logits = q_model(images)
            fp_logits = fp_model(images)

            keys = hash_key(q_logits, topk=topk, num_bins=num_bins)
            for i, key in enumerate(keys):
                correction = fp_logits[i] - q_logits[i]
                lut[key].append(correction.cpu())

    # Average correction per key
    lut_avg = {k: torch.stack(v).mean(0) for k, v in lut.items()}
    return lut_avg
```

```python
def apply_lut(q_logits, lut_avg, topk=5, num_bins=64):
    corrected = []
    keys = hash_key(q_logits, topk=topk, num_bins=num_bins)

    for i, key in enumerate(keys):
        q = q_logits[i]
        if key in lut_avg:
            correction = lut_avg[key].to(q.device)
            corrected.append(q + correction)
        else:
            corrected.append(q)  # fallback: no correction
    return torch.stack(corrected)
```

```python
def evaluate(model, lut_avg, dataloader, device):
    total_top1, total_top5, total = 0, 0, 0

    model.eval()
    with torch.no_grad():
        for i, (images, targets) in enumerate(dataloader):
            if i>=8192:
                break
            
            images, targets = images.to(device), targets.to(device)
            q_logits = model(images)
            corrected_logits = apply_lut(q_logits, lut_avg, topk=3, num_bins=64)

            acc1, acc5 = accuracy(corrected_logits, targets, topk=(1, 5))
            batch_size = images.size(0)
            total_top1 += acc1.item() * batch_size
            total_top5 += acc5.item() * batch_size
            total += batch_size

    print(f"Top-1 Accuracy: {total_top1 / total:.2f}%")
    print(f"Top-5 Accuracy: {total_top5 / total:.2f}%")
```

```python
def build_cluster_lut(q_model, fp_model, dataloader, device, 
                      num_clusters=64, pca_dim=None):
    q_model.eval()
    fp_model.eval()

    all_q, all_fp = [], []

    with torch.no_grad():
        for images, _ in dataloader:
            images = images.to(device)
            q_logits = q_model(images)
            fp_logits = fp_model(images)
            all_q.append(q_logits.cpu())
            all_fp.append(fp_logits.cpu())

    all_q = torch.cat(all_q, dim=0)  # [N, C]
    all_fp = torch.cat(all_fp, dim=0)

    # Optional PCA for speed/denoising
    if pca_dim is not None and pca_dim < all_q.shape[1]:
        pca = PCA(n_components=pca_dim, random_state=42)
        q_features = pca.fit_transform(all_q.numpy())
    else:
        q_features = all_q.numpy()

    # Clustering (KMeans here, can swap for GMM)
    cluster_model = KMeans(n_clusters=num_clusters, random_state=42)
    cluster_ids = cluster_model.fit_predict(q_features)

    # Build correction LUT: cluster_id -> mean(fp - q)
    lut = defaultdict(list)
    for i, cid in enumerate(cluster_ids):
        correction = all_fp[i] - all_q[i]
        lut[int(cid)].append(correction)

    lut_avg = {cid: torch.stack(v).mean(0) for cid, v in lut.items()}

    return cluster_model, lut_avg
```

```python
def apply_cluster_lut(q_logits, cluster_model, lut_avg, pca=None):
    q_np = q_logits.cpu().numpy()

    # Apply PCA if used during build
    if pca is not None:
        q_np = pca.transform(q_np)

    cluster_ids = cluster_model.predict(q_np)

    corrected = []
    for i, q in enumerate(q_logits):
        cid = int(cluster_ids[i])
        if cid in lut_avg:
            correction = lut_avg[cid].to(q.device)
            corrected.append(q + correction)
        else:
            corrected.append(q)
    return torch.stack(corrected)
```

```python
def evaluate_with_lut(q_model, cluster_model, lut_avg, dataloader, device, pca=None):
    total_top1, total_top5, total = 0, 0, 0

    q_model.eval()
    with torch.no_grad():
        for images, targets in dataloader:
            images, targets = images.to(device), targets.to(device)
            q_logits = q_model(images)
            corrected_logits = apply_cluster_lut(q_logits, cluster_model, lut_avg, pca=pca)

            acc1, acc5 = accuracy(corrected_logits, targets, topk=(1, 5))
            batch_size = images.size(0)
            total_top1 += acc1.item() * batch_size
            total_top5 += acc5.item() * batch_size
            total += batch_size

    print(f"Top-1 Accuracy: {total_top1 / total:.2f}%")
    print(f"Top-5 Accuracy: {total_top5 / total:.2f}%")
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
from collections import defaultdict
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
```

```python
def accuracy(output, target, topk=(1,)):
    with torch.no_grad():
        maxk = max(topk)
        batch_size = target.size(0)

        _, pred = output.topk(maxk, 1, True, True)
        pred = pred.t()
        correct = pred.eq(target.view(1, -1).expand_as(pred))

        res = []
        for k in topk:
            correct_k = correct[:k].reshape(-1).float().sum(0, keepdim=True)
            res.append(correct_k.mul_(100.0 / batch_size))
        return res
```

```python
class LearnableCorrectionScaler(nn.Module):
    def __init__(self):
        super().__init__()
        self.w = nn.Parameter(torch.tensor(1.0))
        self.b = nn.Parameter(torch.tensor(0.0))

    def forward(self, q_logits, fp_logits, lut_correction):
        # Get top-5 indices
        q_top5 = torch.topk(q_logits, k=5, dim=-1).indices  # [B, 5]
        fp_top5 = torch.topk(fp_logits, k=5, dim=-1).indices  # [B, 5]

        # Compute agreement score: fraction of FP's top-5 found in Q's top-5
        matches = (q_top5.unsqueeze(2) == fp_top5.unsqueeze(1)).any(dim=1).float()
        agreement_score = matches.sum(dim=1) / 5.0  # [B]

        # Scale factor (learnable)
        scale = torch.sigmoid(self.w * agreement_score + self.b).unsqueeze(1)  # [B, 1]

        # Apply scaled correction
        corrected_logits = q_logits + scale * lut_correction
        return corrected_logits, scale
```

```python
def build_cluster_lut(q_model, fp_model, q_logit,fp_logit, device, num_clusters=64, pca_dim=None):
    q_model.eval()
    fp_model.eval()

    all_q, all_fp = q_logit,fp_logit

    

    
    # Optional PCA
    pca = None
    if pca_dim is not None and pca_dim < all_q.shape[1]:
        pca = PCA(n_components=pca_dim, random_state=42)
        q_features = pca.fit_transform(all_q.numpy())
    else:
        q_features = all_q.numpy()

    # Clustering
    cluster_model = KMeans(n_clusters=num_clusters, random_state=42)
    cluster_ids = cluster_model.fit_predict(q_features)

    # LUT: cluster_id -> mean(fp - q)
    lut = defaultdict(list)
    for i, cid in enumerate(cluster_ids):
        correction = all_fp[i] - all_q[i]
        lut[int(cid)].append(correction)

    lut_avg = {cid: torch.stack(v).mean(0) for cid, v in lut.items()}
    return cluster_model, lut_avg, pca
```

```python
def apply_cluster_lut(q_logits, cluster_model, lut_avg, pca=None):
    q_np = q_logits.cpu().numpy()
    if pca is not None:
        q_np = pca.transform(q_np)
    cluster_ids = cluster_model.predict(q_np)

    corrections = []
    for i, q in enumerate(q_logits):
        cid = int(cluster_ids[i])
        if cid in lut_avg:
            corrections.append(lut_avg[cid].to(q.device))
        else:
            corrections.append(torch.zeros_like(q))
    return torch.stack(corrections)
```

```python
def train_scaler(scaler, q_model, fp_model, cluster_model, lut_avg, dataloader, device, pca=None, epochs=5, lr=1e-2):
    optimizer = torch.optim.Adam(scaler.parameters(), lr=lr)
    scaler.train()

    for epoch in range(epochs):
        total_loss = 0
        for images, _ in dataloader:
            images = images.to(device)

            with torch.no_grad():
                q_logits = q_model(images)
                fp_logits = fp_model(images)
                lut_correction = apply_cluster_lut(q_logits, cluster_model, lut_avg, pca=pca)

            corrected_logits, _ = scaler(q_logits, fp_logits, lut_correction)

            # Loss: align corrected logits with FP logits
            loss = F.mse_loss(corrected_logits, fp_logits)

            optimizer.zero_grad()
            loss.backward()
            optimizer.step()

            total_loss += loss.item() * images.size(0)

        print(f"Epoch {epoch+1}/{epochs}, Loss: {total_loss / len(dataloader.dataset):.6f}")
```

```python
def evaluate_with_scaler(q_model, fp_model, scaler, cluster_model, lut_avg, dataloader, device, pca=None):
    scaler.eval()
    total_top1, total_top5, total = 0, 0, 0

    with torch.no_grad():
        for i,(images, targets) in enumerate(dataloader):
            if i>=512:
                break
            images, targets = images.to(device), targets.to(device)
            q_logits = q_model(images)
            fp_logits = fp_model(images)
            lut_correction = apply_cluster_lut(q_logits, cluster_model, lut_avg, pca=pca)

            corrected_logits, scale = scaler(q_logits, fp_logits, lut_correction)

            acc1, acc5 = accuracy(corrected_logits, targets, topk=(1, 5))
            total_top1 += acc1.item() * images.size(0)
            total_top5 += acc5.item() * images.size(0)
            total += images.size(0)

    print(f"Top-1 Accuracy: {total_top1 / total:.2f}%")
    print(f"Top-5 Accuracy: {total_top5 / total:.2f}%")
```

```python
def evaluate_with_agreement_weights(q_model, fp_model, cluster_model, lut_avg, dataloader, device, pca=None, bucket_weights=[1.0, 0.8, 0.6, 0.4, 0.2]):
    """
    bucket_weights: list of 5 weights for agreement levels 0..4/5
    """
    q_model.eval()
    total_top1, total_top5, total = 0, 0, 0

    with torch.no_grad():
        for images, targets in dataloader:
            images, targets = images.to(device), targets.to(device)

            q_logits = q_model(images)
            fp_logits = fp_model(images)
            lut_correction = apply_cluster_lut(q_logits, cluster_model, lut_avg, pca=pca)

            # Compute agreement level
            q_top5 = torch.topk(q_logits, k=5, dim=-1).indices
            fp_top5 = torch.topk(fp_logits, k=5, dim=-1).indices
            matches = (q_top5.unsqueeze(2) == fp_top5.unsqueeze(1)).any(dim=1).float()
            agreement_count = matches.sum(dim=1).long()  # [B], 0..5
            agreement_bucket = torch.clamp(agreement_count, max=4)  # merge 4 & 5 into one bucket

            # Apply per-bucket weights
            weights = torch.tensor(bucket_weights, device=device)[agreement_bucket].unsqueeze(1)
            corrected_logits = q_logits + weights * lut_correction

            acc1, acc5 = accuracy(corrected_logits, targets, topk=(1, 5))
            total_top1 += acc1.item() * images.size(0)
            total_top5 += acc5.item() * images.size(0)
            total += images.size(0)

    print(f"Top-1 Accuracy: {total_top1 / total:.2f}%")
    print(f"Top-5 Accuracy: {total_top5 / total:.2f}%")
```

```python
import torch
import torch.nn as nn
from collections import defaultdict
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans

# -----------------------
# Accuracy computation
# -----------------------
def accuracy(output, target, topk=(1,)):
    with torch.no_grad():
        maxk = max(topk)
        batch_size = target.size(0)

        _, pred = output.topk(maxk, 1, True, True)
        pred = pred.t()
        correct = pred.eq(target.view(1, -1).expand_as(pred))

        res = []
        for k in topk:
            correct_k = correct[:k].reshape(-1).float().sum(0, keepdim=True)
            res.append(correct_k.mul_(100.0 / batch_size))
        return res

# -----------------------
# Build per-cluster affine parameters
# -----------------------
def build_cluster_affine(q_model, fp_model, dataloader, device, num_clusters=64, pca_dim=None):
    q_model.eval()
    fp_model.eval()

    all_q, all_fp = [], []

    with torch.no_grad():
        for images, _ in dataloader:
            images = images.to(device)
            q_logits = q_model(images)
            fp_logits = fp_model(images)
            all_q.append(q_logits.cpu())
            all_fp.append(fp_logits.cpu())

    all_q = torch.cat(all_q, dim=0)  # [N, C]
    all_fp = torch.cat(all_fp, dim=0)  # [N, C]

    # Optional PCA for clustering only
    pca = None
    if pca_dim is not None and pca_dim < all_q.shape[1]:
        pca = PCA(n_components=pca_dim, random_state=42)
        q_features = pca.fit_transform(all_q.numpy())
    else:
        q_features = all_q.numpy()

    # Cluster quantized outputs
    cluster_model = KMeans(n_clusters=num_clusters, random_state=42)
    cluster_ids = cluster_model.fit_predict(q_features)

    # For each cluster: learn gamma, beta (per-class)
    gamma_dict = {}
    beta_dict = {}

    for cid in range(num_clusters):
        idxs = (cluster_ids == cid)
        if idxs.sum() == 0:
            # Empty cluster, default to identity
            gamma_dict[cid] = torch.ones(all_q.shape[1])
            beta_dict[cid] = torch.zeros(all_q.shape[1])
            continue

        q_c = all_q[idxs]  # [Nc, C]
        fp_c = all_fp[idxs]  # [Nc, C]

        # Closed-form least squares: fp ≈ gamma * q + beta
        mean_q = q_c.mean(dim=0)
        mean_fp = fp_c.mean(dim=0)

        # Compute variance, avoid div by zero
        var_q = q_c.var(dim=0, unbiased=False)
        var_q[var_q < 1e-8] = 1e-8

        gamma = ((q_c - mean_q) * (fp_c - mean_fp)).mean(dim=0) / var_q
        beta = mean_fp - gamma * mean_q

        gamma_dict[cid] = gamma
        beta_dict[cid] = beta

    return cluster_model, gamma_dict, beta_dict, pca

# -----------------------
# Apply per-cluster affine correction
# -----------------------
def apply_cluster_affine(q_logits, cluster_model, gamma_dict, beta_dict, pca=None):
    q_np = q_logits.cpu().numpy()
    if pca is not None:
        q_np = pca.transform(q_np)
    cluster_ids = cluster_model.predict(q_np)

    corrected = []
    for i, q in enumerate(q_logits):
        cid = int(cluster_ids[i])
        gamma = gamma_dict[cid].to(q.device)
        beta = beta_dict[cid].to(q.device)
        affine_corrected = q * gamma + beta
        # Blend with original q_logits
        blended = q + alpha * (affine_corrected - q)
        corrected.append(blended)
    return torch.stack(corrected)

# -----------------------
# Evaluate
# -----------------------
def evaluate_cluster_affine(q_model, cluster_model, gamma_dict, beta_dict, dataloader, device, pca=None):
    q_model.eval()
    total_top1, total_top5, total = 0, 0, 0

    with torch.no_grad():
        for images, targets in dataloader:
            images, targets = images.to(device), targets.to(device)
            q_logits = q_model(images)
            corrected_logits = apply_cluster_affine(q_logits, cluster_model, gamma_dict, beta_dict, pca=pca)

            acc1, acc5 = accuracy(corrected_logits, targets, topk=(1, 5))
            total_top1 += acc1.item() * images.size(0)
            total_top5 += acc5.item() * images.size(0)
            total += images.size(0)

    print(f"Top-1 Accuracy: {total_top1 / total:.2f}%")
    print(f"Top-5 Accuracy: {total_top5 / total:.2f}%")

# -----------------------
# Example usage
# -----------------------
# cluster_model, gamma_dict, beta_dict, pca = build_cluster_affine(q_model, fp_model, train_loader, device, num_clusters=64, pca_dim=50)
# print("Validation Accuracy with Per-Cluster Affine Correction:")
# evaluate_cluster_affine(q_model, cluster_model, gamma_dict, beta_dict, val_loader, device, pca=pca)
```

```python
def monitor_and_plot_errors(q_logits, fp_logits, metric="mse"):
    # Compute per-row error
    if metric == "mse":
        row_errors = F.mse_loss(q_logits, fp_logits, reduction='none').mean(dim=1)  # [N]
    elif metric == "l1":
        row_errors = F.l1_loss(q_logits, fp_logits, reduction='none').mean(dim=1)
    else:
        raise ValueError("metric must be 'mse' or 'l1'")

    # Accumulate errors over rows
    accumulated_error = torch.cumsum(row_errors, dim=0)  # [N]

    # Plot
    plt.figure(figsize=(10,5))
    plt.plot(accumulated_error.numpy())
    plt.xlabel("Row index")
    plt.ylabel(f"Accumulated {metric.upper()} Error")
    plt.title(f"Accumulated {metric.upper()} Error over Samples")
    plt.grid(True)
    plt.show()

    return row_errors, accumulated_error
```

```python
def average_error_per_logit(q_logits, fp_logits, metric="mse"):
    """
    q_logits, fp_logits: [N, C] tensors
    Returns: [C] average error for each logit index
    """
    if metric == "mse":
        errors = F.mse_loss(q_logits, fp_logits, reduction='none')  # [N, C]
    elif metric == "l1":
        errors = F.l1_loss(q_logits, fp_logits, reduction='none')
    else:
        raise ValueError("metric must be 'mse' or 'l1'")

    avg_error_per_logit = errors.mean(dim=0)  # [C]
    return avg_error_per_logit

def plot_avg_error(avg_error_per_logit, metric="mse"):
    plt.figure(figsize=(12,5))
    plt.plot(avg_error_per_logit.numpy())
    plt.xlabel("Logit index (class)")
    plt.ylabel(f"Average {metric.upper()} Error")
    plt.title(f"Average {metric.upper()} Error per Logit Position")
    plt.grid(True)
    plt.show()
```

```python
def compute_transformed_class_errors(q_logits, fp_logits):
    """
    q_logits, fp_logits: [N, C]
    Returns: transformed_errors [C]
    """
    mse_per_class = ((q_logits - fp_logits) ** 2).mean(dim=0)  # [C]
    transformed = torch.log(mse_per_class + 1e-8) ** 2
    return transformed
```

```python
def cluster_classes_by_error(transformed_errors, num_clusters=8):
    """
    transformed_errors: [C] tensor
    Returns: cluster_model, cluster_ids [C]
    """
    features = transformed_errors.unsqueeze(1).numpy()  # [C, 1]
    cluster_model = KMeans(n_clusters=num_clusters, random_state=42)
    cluster_ids = cluster_model.fit_predict(features)
    return cluster_model, cluster_ids
```

```python
def assign_group_strengths(cluster_ids, base_strength=0.5):
    """
    Assign correction strengths inversely proportional to cluster index.
    More sensitive groups get higher strength.
    """
    num_clusters = cluster_ids.max() + 1
    strengths = {}
    for cid in range(num_clusters):
        # Example heuristic: highest error cluster gets base_strength*2
        strengths[cid] = base_strength + (num_clusters - cid - 1) * (base_strength / num_clusters)
    return strengths
```

```python
def apply_grouped_correction(q_logits, lut_correction, cluster_ids, strengths):
    """
    Apply correction scaled by group strength per class.
    q_logits, lut_correction: [B, C]
    cluster_ids: [C]
    strengths: dict cluster_id -> float
    """
    scales = torch.tensor([strengths[cid] for cid in cluster_ids], device=q_logits.device)  # [C]
    return q_logits + lut_correction * scales.unsqueeze(0)  # [B, C]
```

```python
import torch
import torch.nn as nn
from sklearn.cluster import KMeans
import numpy as np

# -----------------------
# Accuracy computation
# -----------------------
def accuracy(output, target, topk=(1,)):
    with torch.no_grad():
        maxk = max(topk)
        batch_size = target.size(0)
        _, pred = output.topk(maxk, 1, True, True)
        pred = pred.t()
        correct = pred.eq(target.view(1, -1).expand_as(pred))
        res = []
        for k in topk:
            correct_k = correct[:k].reshape(-1).float().sum(0, keepdim=True)
            res.append(correct_k.mul_(100.0 / batch_size))
        return res

# -----------------------
# Build per-class-group affine parameters (log error space)
# -----------------------
def build_cluster_affine_log_error(q_model, fp_model, dataloader, device, num_clusters=8):
    q_model.eval()
    fp_model.eval()

    all_q, all_fp = [], []

    with torch.no_grad():
        for images, _ in dataloader:
            images = images.to(device)
            q_logits = q_model(images)
            fp_logits = fp_model(images)
            all_q.append(q_logits.cpu())
            all_fp.append(fp_logits.cpu())

    all_q = torch.cat(all_q, dim=0)  # [N, C]
    all_fp = torch.cat(all_fp, dim=0)  # [N, C]
    C = all_q.shape[1]

    # Step 1: Compute per-class mean error
    mse_per_class = ((all_q - all_fp) ** 2).mean(dim=0)  # [C]
    transformed_error = torch.log(mse_per_class + 1e-8) ** 2  # [C]

    # Step 2: Cluster classes based on transformed error
    cluster_model = KMeans(n_clusters=num_clusters, random_state=42)
    class_cluster_ids = cluster_model.fit_predict(transformed_error.unsqueeze(1).numpy())

    # Step 3: Learn gamma/beta per cluster
    gamma_dict = {}
    beta_dict = {}
    for cid in range(num_clusters):
        idxs = np.where(class_cluster_ids == cid)[0]  # class indices
        if len(idxs) == 0:
            gamma_dict[cid] = torch.ones(C)
            beta_dict[cid] = torch.zeros(C)
            continue

        q_c = all_q[:, idxs]  # [N, |idxs|]
        fp_c = all_fp[:, idxs]

        mean_q = q_c.mean(dim=0)
        mean_fp = fp_c.mean(dim=0)
        var_q = q_c.var(dim=0, unbiased=False)
        var_q[var_q < 1e-8] = 1e-8

        gamma = ((q_c - mean_q) * (fp_c - mean_fp)).mean(dim=0) / var_q
        beta = mean_fp - gamma * mean_q

        # Store gamma/beta for each class in cluster
        gamma_full = torch.ones(C)
        beta_full = torch.zeros(C)
        for j, cls_idx in enumerate(idxs):
            gamma_full[cls_idx] = gamma[j]
            beta_full[cls_idx] = beta[j]

        gamma_dict[cid] = gamma_full
        beta_dict[cid] = beta_full

    return cluster_model, class_cluster_ids, gamma_dict, beta_dict

# -----------------------
# Apply per-class-group affine correction
# -----------------------
def apply_cluster_affine_log_error(q_logits, cluster_model, class_cluster_ids, gamma_dict, beta_dict, alpha=1.0):
    C = q_logits.shape[1]
    corrected = []
    for q in q_logits:
        new_q = q.clone()
        for c in range(C):
            cid = int(class_cluster_ids[c])
            gamma = gamma_dict[cid][c].to(q.device)
            beta = beta_dict[cid][c].to(q.device)
            affine_corrected = q[c] * gamma + beta
            new_q[c] = q[c] + alpha * (affine_corrected - q[c])
        corrected.append(new_q)
    return torch.stack(corrected)

# -----------------------
# Evaluate
# -----------------------
def evaluate_cluster_affine_log_error(q_model, cluster_model, class_cluster_ids, gamma_dict, beta_dict, dataloader, device, alpha=1.0):
    q_model.eval()
    total_top1, total_top5, total = 0, 0, 0

    with torch.no_grad():
        for images, targets in dataloader:
            images, targets = images.to(device), targets.to(device)
            q_logits = q_model(images)
            corrected_logits = apply_cluster_affine_log_error(q_logits, cluster_model, class_cluster_ids, gamma_dict, beta_dict, alpha=alpha)
            acc1, acc5 = accuracy(corrected_logits, targets, topk=(1, 5))
            total_top1 += acc1.item() * images.size(0)
            total_top5 += acc5.item() * images.size(0)
            total += images.size(0)

    print(f"Top-1 Accuracy: {total_top1 / total:.2f}%")
    print(f"Top-5 Accuracy: {total_top5 / total:.2f}%")
```

```python
cluster_model, class_cluster_ids, gamma_dict, beta_dict = build_cluster_affine_log_error(
    qnn, fp_model, train_loader, device,
    num_clusters=16  # tune: fewer clusters => less overfit risk
)
```

```python
print("\nValidation Accuracy with Log-Error-Based Affine Correction:")
evaluate_cluster_affine_log_error(
    qnn, cluster_model, class_cluster_ids, gamma_dict, beta_dict,
    test_loader, device,
    alpha=0.5  # blend factor to avoid overcorrection, tune between 0.3–1.0
)
```
